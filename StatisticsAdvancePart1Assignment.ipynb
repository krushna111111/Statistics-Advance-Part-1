#1 What is a random variable in probability theory?

   Q1-Answer - In probability theory, a random variable is essentially a way to assign numerical values to the outcomes of a random experiment.

**Random Experiment:**

This is any process whose outcome is uncertain. Examples include flipping a coin, rolling a die, or measuring the height of a person.   


**Sample Space:**

This is the set of all possible outcomes of a random experiment. For example, the sample space for flipping a coin is {heads, tails}, and for rolling a six-sided die, it's {1, 2, 3, 4, 5, 6}.   


**Random Variable as a Function:**

A random variable is a function that maps the outcomes in the sample space to real numbers.
 This allows us to work with these outcomes mathematically.

**Types of Random Variables:**

**Discrete Random Variables:** These can only take on a countable number of values. For example, the number of heads in a series of coin flips, or the number of cars passing a certain point on a road in an hour.   

**Continuous Random Variables:** These can take on any value within a given range.
 For example, the height of a person, or the temperature of a room.

**Purpose:**

Random variables provide a way to quantify and analyze random phenomena.
 They allow us to use the tools of mathematics and statistics to understand and predict the likelihood of different outcomes.

#11 What are the characteristics of a normal distribution?

   Q11-Answer- The normal distribution, also known as the Gaussian distribution, is a fundamental concept in statistics.It's characterized by its distinctive bell-shaped curve and several key properties:

**Bell-Shaped Curve:**

  The graph of a normal distribution forms a symmetrical, bell-shaped curve.   


**Symmetry:**

  The distribution is perfectly symmetrical around its mean. This means that the left and right sides of the curve are mirror images of each other.   


**Mean, Median, and Mode are Equal:**

  In a normal distribution, the mean (average), median (middle value), and mode (most frequent value) are all the same. They are located at the center of the distribution.   


   
**Defined by Mean and Standard Deviation:**

  The normal distribution is completely defined by two parameters:

**Mean (μ):** Determines the center of the distribution.   

**Standard deviation (σ):** Determines the spread or width of the
  distribution.   


   


**Empirical Rule (68-95-99.7 Rule):**

**This rule describes the proportion of data that falls within certain standard deviations of the mean:**

  Approximately 68% of the data falls within one standard deviation of the mean.   

  Approximately 95% of the data falls within two standard deviations of the mean.   

  Approximately 99.7% of the data falls within three standard deviations of the mean.   

   



**Asymptotic Tails:**

  The tails of the normal distribution extend infinitely in both directions, approaching the horizontal axis but never touching it.   


**Total Area Under the Curve:**

  The total area under the normal distribution curve is equal to 1, representing the total probability of all possible outcomes.

#13 What is the Central Limit Theorem (CLT), and why is it critical in statistics?

   Q13-Answer- The Central Limit Theorem (CLT) is a cornerstone of probability theory and statistics.It essentially states that the distribution of sample means approaches a normal distribution as the sample size increases, regardless of the shape of the original population distribution.

**What the Central Limit Theorem Says:**

**Sample Means:**

  Imagine you have a population with any distribution (it could be skewed, uniform, or anything else).   

  If you take multiple random samples from that population and calculate the mean of each sample, those sample means will form a new distribution.   


**Approaching Normality:**

  As the sample size (n) gets larger, the distribution of these sample means will increasingly resemble a normal distribution.   



**Mean and Standard Deviation:**

  The mean of the sample means will be equal to the population mean (μ).   

  The standard deviation of the sample means (also known as the standard error) will be equal to the population standard deviation (σ) divided by the square root of the sample size (n): σ/√n.   



**Why the Central Limit Theorem is Critical:**

**Statistical Inference:**

  It allows us to make inferences about a population based on sample data, even when we don't know the population's distribution. Many statistical tests and confidence intervals rely on the CLT.   


**Hypothesis Testing:**

  It's essential for hypothesis testing, as it allows us to calculate probabilities and determine whether observed sample means are statistically significant.   



**Confidence Intervals:**

  It's used to construct confidence intervals, which provide a range of values within which the population mean is likely to fall.


**Simplifying Complex Data:**

  It simplifies the analysis of complex data by allowing us to work with the normal distribution, which is well-understood and has many convenient properties.   


**Real-World Applications:**

  It's applicable in various fields, including economics, medicine, engineering, and social sciences, where we often deal with sample data to draw conclusions about populations.


**Enables the use of Z-scores:**

  Because of the CLT, we can use z-scores and the standard normal distribution to calculate probabilities of sample means, even when the original population is not normally distributed.

#15 What is the application of Z statistics in hypothesis testing?

**Core Function:**

**Measuring Statistical Significance:**

  The Z-statistic quantifies how many standard deviations a sample mean deviates from the population mean. This allows us to determine if the observed difference is statistically significant or likely due to random chance.   



**Hypothesis Testing Framework:**

  In hypothesis testing, we formulate a null hypothesis (H0) and an alternative hypothesis (H1). The Z-statistic helps us decide whether to reject the null hypothesis in favor of the alternative.   



**Key Applications:**

**Comparing Means:**

**One-sample Z-test:**

  Used to determine if the mean of a single sample differs significantly from a known population mean.   



**Two-sample Z-test:**

  Used to compare the means of two independent samples to see if they are significantly different.   




**Conditions for Use:**

**Z-tests are most appropriate when:**

  The population standard deviation is known.   

  The sample size is large (typically n ≥ 30), due to the Central Limit Theorem, which ensures the distribution of sample means approaches normality.   

  The data is normally distributed.   





**Decision Making:**

  By calculating the Z-statistic, we can compare it to critical values from the standard normal distribution.   

  If the Z-statistic falls within the critical region (determined by the chosen significance level), we reject the null hypothesis.   

  This process allows us to make informed decisions based on statistical evidence.   



**Examples:**

**Quality control:** Testing if the average weight of products from a
  manufacturing line meets specifications.

#16 How do you calculate a Z-score, and what does it represent?

   Q16-Answer- Calculating a Z-score is a straightforward process, and it provides valuable information about a data point's position within a distribution.

  Calculating a Z-score:

  The formula for calculating a Z-score is:

Z= X−μ /  σ

**Where:**

**Z:** The Z-score.   

**X:** The individual data point.   

**μ (mu):** The mean of the population or sample.   

**σ (sigma):** The standard deviation of the population or sample.   

**What a Z-score Represents:**

**Standard Deviations from the Mean:**

A Z-score tells you how many standard deviations a particular data point is away from the mean of the distribution.   



**Positive Z-score:**

A positive Z-score indicates that the data point is above the mean.   


**Negative Z-score:**

A negative Z-score indicates that the data point is below the mean.   



**Z-score of Zero:**

A Z-score of 0 means that the data point is exactly equal to the mean.   


**Standardization:**

Z-scores standardize data, allowing you to compare data points from different distributions. By converting data to Z-scores, you can see how each point relates to its respective mean and standard deviation on a common scale.   



**Probability and Percentiles:**

Z-scores are used to find probabilities and percentiles associated with data points in a normal distribution.

Using a Z-table (standard normal distribution table), you can determine the proportion of data that falls above or below a specific Z-score.   



**Example:**

Let's say you have a test score of 80, the mean of the test scores is 70, and the standard deviation is 5.

**To calculate the Z-score:**

Z = 80−70 / 5​ = 10/5 ​= 2

This Z-score of 2 indicates that your test score is 2 standard deviations above the mean.

#17 What are point estimates and interval estimates in statistics?

   Q17-Answer- In statistics, when we want to estimate population parameters (like the mean or proportion) using sample data, we use two main types of estimates: point estimates and interval estimates.

**Point Estimates:**

**Definition:**

  A point estimate is a single value that serves as the "best guess" or estimate of an unknown population parameter.   



**Examples:**

  The sample mean (x̄) is a point estimate of the population mean (μ).   

  The sample proportion (p̂) is a point estimate of the population proportion (p).   



**Characteristics:**

  It's a straightforward, single number.   

  While simple, it doesn't provide any information about the precision or reliability of the estimate.   



**Limitations:**

  It's highly unlikely that a point estimate will exactly equal the true population parameter.   

  It doesn't convey the uncertainty associated with the estimate.   



**Interval Estimates:**

**Definition:**

  An interval estimate provides a range of values within which the population parameter is likely to fall, along with a level of confidence that the parameter lies within that range.   



**Examples:**

  Confidence intervals are the most common type of interval estimate.   



**Characteristics:**

  It provides a range (e.g., "the population mean is likely between 10 and 12").   

  It includes a confidence level (e.g., "we are 95% confident that the population mean is within this range").   

  It conveys the uncertainty associated with the estimate.   


**Advantages:**

  It provides a more informative estimate than a single point estimate.   

  It reflects the variability of the sample data.

  It allows us to express the reliability of our estimate.

#18 What is the significance of confidence intervals in statistical analysis?

   Q18-Answer- Confidence intervals are a fundamental tool in statistical analysis, and their significance stems from their ability to provide a more informative and reliable way to estimate population parameters than simple point estimates.

**Key Significance:**

**Quantifying Uncertainty:**

  Confidence intervals acknowledge that sample data is subject to variability and that point estimates are unlikely to be perfectly accurate. They provide a range of values that are likely to contain the true population parameter, along with a level of confidence associated with that range.   



**Providing a Range of Plausible Values:**

  Instead of offering a single "best guess," confidence intervals give a range of plausible values for the population parameter. This range reflects the uncertainty inherent in using sample data to make inferences about the population.   


**Expressing Reliability:**

  The confidence level (e.g., 95%, 99%) indicates the reliability of the interval. A 95% confidence interval, for instance, means that if we were to repeat the sampling process many times, 95% of the resulting intervals would contain the true population parameter.   



**Facilitating Decision-Making:**

  Confidence intervals help in making informed decisions by providing a sense of the precision of the estimate. If the interval is narrow, it suggests a more precise estimate. If it's wide, it indicates greater uncertainty.   


**Hypothesis Testing Connection:**

  Confidence intervals are related to hypothesis testing. If the hypothesized value of the population parameter falls outside the confidence interval, it provides evidence to reject the null hypothesis.   



**Practical Implications:**

  In various fields, confidence intervals are crucial for interpreting research findings and making practical
  decisions. For example:

  In medical research, they help determine the effectiveness of a new treatment.   

  In market research, they help estimate consumer preferences.   

  In quality control, they help assess the variability of product characteristics.   

   



**Understanding Statistical Significance:**

  While p values are also used to determine statistical significance, confidence intervals display the range of plausible values that the true population parameter could be. This gives a more practical understanding of the data.

**Key Relationship:**

**Z-scores as the Foundation:**

  When constructing confidence intervals for population means (when the population standard deviation is known or with large samples), we utilize Z-scores from the standard normal distribution.

**Confidence Level and Z-score:**

  The desired confidence level (e.g., 95%, 99%) dictates the corresponding Z-score.

  The basic formula for a confidence interval involving Z-scores is:

  Sample Mean ± (Z-score * Standard Error)


  Where the standard error is the standard deviation of the sample mean.


**How They Work Together:**

  Essentially, the Z-score provides the "multiplier" that determines how far we extend the interval around the sample mean.

  The standard error reflects the variability of the sample mean.   

  By combining the Z-score and the standard error, we create a range of values within which we are confident the population mean lies.

#21 What are the assumptions for applying the Central Limit Theorem?

   Q21-Answer- The Central Limit Theorem (CLT) is a powerful tool, but it relies on certain assumptions to hold true.

**Independence:**

  The samples must be drawn independently of each other.

  This means that the value of one sample should not influence the value of any other sample.   

  This assumption is crucial for ensuring that the sample means are not biased.

**Sample Size (n):**

  While the CLT applies to any population distribution, the sample size (n) plays a critical role in how quickly the distribution of sample means approaches normality.   

  A sufficiently large sample size is required. While there's no strict rule, a common guideline is that n ≥ 30 is generally considered large enough for the CLT to hold reasonably well.   

  However, if the population distribution is already close to normal, the CLT can work with smaller sample sizes.

  If the original population is heavily skewed, a larger sample size may be necessary.   

**Finite Variance:**

  The population from which the samples are drawn must have a finite variance (σ²). This ensures that the standard error of the sample means (σ/√n) is well-defined.   

  Distributions with infinite variance, such as the Cauchy distribution, do not satisfy this assumption, and the CLT does not apply.   

**Important Considerations:**

**Random Sampling:** The samples must be drawn randomly from the
  population. Any bias in the sampling process can invalidate the CLT.   

**Population Distribution:** While the CLT holds regardless
  of the population's distribution, the shape of the population distribution can affect how quickly the distribution of sample means approaches normality.

  Highly skewed distributions may require larger sample sizes.

#22 What is the concept of expected value in a probability distribution?

   Q22-Answer- The concept of expected value, often denoted as E(X) or μ (mu), is fundamental in probability theory and statistics.It represents the long-run average or mean of a random variable.

**Core Idea:**

**Weighted Average:**

  The expected value is essentially a weighted average of all possible values that a random variable can take.   

  The weights are the probabilities associated with each value.   


**Long-Run Average:**

  If you were to repeat a random experiment many times and calculate the average of the outcomes, that average would approach the expected value.


**Central Tendency:**

  The expected value provides a measure of the central tendency of a probability distribution.   



**Calculation:**

**Discrete Random Variable:**

  For a discrete random variable X, the expected value is
  calculated as:

E(X) = Σ [x * P(x)]

**Where:**

  x represents each possible value of X.   

  P(x) represents the probability of X taking the value x.   

  Σ denotes the sum of all possible values.






**Continuous Random Variable:**

  For a continuous random variable X, the expected value is calculated as:

  E(X) = ∫ [x * f(x)] dx

**Where:**

  f(x) is the probability density function of X.

  ∫ denotes the integral over all possible values of X.






**Applications and Significance:**

**Decision-Making:**

  The expected value is crucial in decision-making, particularly in situations involving uncertainty. It helps assess the average outcome of different choices.   

  For example, in gambling or investment, the expected value helps determine the long-term profitability or loss.   



**Risk Assessment:**

  It's used in risk assessment to quantify the average potential loss or gain associated with a particular event.   


**Insurance:**

  Insurance companies use expected value to calculate premiums based on the average expected payouts.


**Finance:**

  In finance, it's used to evaluate the expected return on investments.   



**Quality Control:**

  In quality control, it's used to determine the average number of defects in a production process.

**Probability Distribution Defines the Possibilities:**

  The expected value then uses this information to calculate the average outcome you would expect to see over many repetitions of the random experiment.

  The expected value is a property of the probability distribution. Changing the probability distribution will change the expected value.

  The probability distribution contains much more information than the expected value. The expected value is just one characteristic of the distribution.

```python
#3 Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution?

def bernoulli_pdf(x, p):
    """
    Calculates the probability density function (PDF) of a Bernoulli distribution.

    Note: While technically Bernoulli is a PMF not PDF, this function is named PDF
    for consistency with common statistical terms.

    Args:
        x (int): The value (0 or 1) for which to calculate the probability.
        p (float): The probability of success (1).

    Returns:
        float: The probability of x.
    """
    if not isinstance(x, int) or x not in (0, 1):
        raise ValueError("x must be either 0 or 1.")
    if not 0 <= p <= 1:
        raise ValueError("p must be between 0 and 1.")

    if x == 1:
        return p
    else:  # x == 0
        return 1 - p

# Example usage:
try:
    print(bernoulli_pdf(1, 0.7))  # Probability of success (1) with p = 0.7
    print(bernoulli_pdf(0, 0.7))  # Probability of failure (0) with p = 0.7
    print(bernoulli_pdf(2, 0.7)) # example error.
    print(bernoulli_pdf(1, 1.5)) # example error.
except ValueError as e:
    print(f"Error: {e}")
```

```python
#6 Q6.Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution.

import matplotlib.pyplot as plt
import numpy as np

def discrete_uniform_cdf(a, b):
    """
    Calculates and plots the cumulative distribution function (CDF) of a discrete uniform distribution.

    Args:
        a (int): The lower bound of the distribution.
        b (int): The upper bound of the distribution.
    """

    if not isinstance(a, int) or not isinstance(b, int):
        raise ValueError("a and b must be integers.")
    if a > b:
        raise ValueError("a must be less than or equal to b.")

    values = np.arange(a, b + 1)
    probabilities = np.full(len(values), 1 / (b - a + 1))  # Equal probabilities for discrete uniform
    cdf = np.cumsum(probabilities)

    plt.step(values, cdf, where='post') #step plot for cdf
    plt.xlabel("Values")
    plt.ylabel("Cumulative Probability")
    plt.title(f"Discrete Uniform CDF (a={a}, b={b})")
    plt.xticks(values)
    plt.grid(True)
    plt.show()

# Example usage:
try:
    discrete_uniform_cdf(1, 6)  # Discrete uniform CDF from 1 to 6
    discrete_uniform_cdf(5, 10) # Discrete uniform CDF from 5 to 10
    discrete_uniform_cdf(10, 5) # Example of error.
    discrete_uniform_cdf(1.5, 6) # Example of error.

except ValueError as e:
    print(f"Error: {e}")
```

```python
#9 Write a Python function to calculate Z-scores from a dataset and plot them.

import matplotlib.pyplot as plt
import numpy as np

def calculate_and_plot_z_scores(data):
    """
    Calculates Z-scores from a dataset and plots them.

    Args:
        data (list or numpy.ndarray): The dataset.
    """

    if not data:
        raise ValueError("Data cannot be empty.")

    data_array = np.array(data)  # Convert to NumPy array for easier calculations
    mean = np.mean(data_array)
    std_dev = np.std(data_array)

    if std_dev == 0:
        raise ValueError("Standard deviation is zero. Z-scores cannot be calculated.")

    z_scores = (data_array - mean) / std_dev

    plt.figure(figsize=(10, 5))
    plt.scatter(range(len(z_scores)), z_scores)
    plt.xlabel("Data Point Index")
    plt.ylabel("Z-score")
    plt.title("Z-scores of the Dataset")
    plt.grid(True, linestyle='--')
    plt.axhline(0, color='red', linestyle='--') # Add horizontal line at z=0
    plt.show()

    return z_scores

# Example usage:
try:
    data1 = [10, 12, 15, 18, 20, 25, 30]
    z_scores1 = calculate_and_plot_z_scores(data1)
    print("Z-scores:", z_scores1)

    data2 = [5, 5, 5, 5, 5]
    z_scores2 = calculate_and_plot_z_scores(data2) #Example error.

    data3 = []
    z_scores3 = calculate_and_plot_z_scores(data3) #Example error.

except ValueError as e:
    print(f"Error: {e}")
```

```python
#15 Simulate multiple samples from a normal distribution and verify the Central Limit Theorem.

import matplotlib.pyplot as plt
import numpy as np

def verify_clt_normal(mu, sigma, sample_size, num_samples=1000):
    """
    Simulates multiple samples from a normal distribution and verifies the Central Limit Theorem.

    Args:
        mu (float): The mean of the normal distribution.
        sigma (float): The standard deviation of the normal distribution.
        sample_size (int): The size of each sample.
        num_samples (int): The number of samples to generate.
    """

    sample_means = []

    for _ in range(num_samples):
        sample = np.random.normal(mu, sigma, sample_size)
        sample_means.append(np.mean(sample))

    # Plotting
    plt.figure(figsize=(10, 5))

    # Histogram of the sample means
    plt.hist(sample_means, bins=30, density=True, alpha=0.6, color='skyblue', edgecolor='black')

    # Overlay the theoretical normal distribution (for comparison)
    x_values = np.linspace(min(sample_means), max(sample_means), 100)
    theoretical_pdf = (1 / (sigma / np.sqrt(sample_size) * np.sqrt(2 * np.pi))) * \
                      np.exp(-0.5 * ((x_values - mu) / (sigma / np.sqrt(sample_size)))**2)
    plt.plot(x_values, theoretical_pdf, 'r-', linewidth=2, label="Theoretical Normal")

    plt.xlabel("Sample Means")
    plt.ylabel("Probability Density")
    plt.title(f"CLT Verification (Normal Pop., Sample Size = {sample_size})")
    plt.grid(True, linestyle='--')
    plt.legend()
    plt.show()

# Example usage:
try:
    mu = 5
    sigma = 2
    sample_size = 30
    verify_clt_normal(mu, sigma, sample_size)

    sample_size = 5;
    verify_clt_normal(mu, sigma, sample_size) #shows that even with small sample sizes, the CLT holds for a normal distribution.

except ValueError as e:
    print(f"Error: {e}")
```

```python
#17 Generate random variables and calculate their corresponding probabilities using the binomial distribution.

import numpy as np
import scipy.stats as stats

def binomial_random_variables_and_probabilities(n, p, num_variables=10):
    """
    Generates random variables and calculates their corresponding probabilities
    using the binomial distribution.

    Args:
        n (int): The number of trials.
        p (float): The probability of success in each trial.
        num_variables (int): The number of random variables to generate.

    Returns:
        tuple: A tuple containing two lists:
            - random_variables (list): The generated random variables.
            - probabilities (list): The corresponding probabilities.
    """

    if not isinstance(n, int) or n <= 0:
        raise ValueError("n must be a positive integer.")
    if not 0 <= p <= 1:
        raise ValueError("p must be between 0 and 1.")
    if not isinstance(num_variables, int) or num_variables <= 0:
        raise ValueError("num_variables must be a positive integer.")

    random_variables = np.random.binomial(n, p, num_variables)
    probabilities = [stats.binom.pmf(x, n, p) for x in random_variables]

    return random_variables.tolist(), probabilities

# Example usage:
try:
    n = 10
    p = 0.6
    random_vars, probs = binomial_random_variables_and_probabilities(n, p, num_variables=5)

    print("Random Variables:", random_vars)
    print("Probabilities:", probs)

    n = -5 #example error
    random_vars, probs = binomial_random_variables_and_probabilities(n, p, num_variables=5)

except ValueError as e:
    print(f"Error: {e}")
```

```python
#18 Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.

import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

def z_score_comparison(data_point, mean, std_dev):
    """
    Calculates the Z-score for a data point and compares it to a standard normal distribution.

    Args:
        data_point (float): The data point to calculate the Z-score for.
        mean (float): The mean of the distribution.
        std_dev (float): The standard deviation of the distribution.
    """

    if std_dev == 0:
        raise ValueError("Standard deviation cannot be zero.")

    z_score = (data_point - mean) / std_dev

    # Plotting
    x = np.linspace(-4, 4, 1000)  # Range for standard normal
    pdf = stats.norm.pdf(x, 0, 1)  # Standard normal PDF

    plt.figure(figsize=(10, 6))
    plt.plot(x, pdf, 'b-', linewidth=2, label="Standard Normal Distribution")
    plt.axvline(z_score, color='r', linestyle='--', label=f"Z-score: {z_score:.2f}")  # Vertical line at Z-score
    plt.xlabel("Z-values")
    plt.ylabel("Probability Density")
    plt.title(f"Z-score Comparison (Data Point: {data_point})")
    plt.legend()
    plt.grid(True, linestyle='--')
    plt.show()

    return z_score

# Example usage:
try:
    data_point = 18
    mean = 15
    std_dev = 2
    z = z_score_comparison(data_point, mean, std_dev)
    print(f"The Z-score is: {z}")

    data_point = 15
    mean = 15
    std_dev = 0
    z = z_score_comparison(data_point, mean, std_dev) #Example error.

except ValueError as e:
    print(f"Error: {e}")
```

```python
#19 Implement hypothesis testing using Z-statistics for a sample dataset.

import numpy as np
import scipy.stats as stats

def z_test_hypothesis(sample_data, population_mean, population_std, alpha=0.05, alternative='two-sided'):
    """
    Performs a Z-test hypothesis test for a sample dataset.

    Args:
        sample_data (list or numpy.ndarray): The sample dataset.
        population_mean (float): The hypothesized population mean.
        population_std (float): The population standard deviation.
        alpha (float): The significance level (default: 0.05).
        alternative (str): The alternative hypothesis ('two-sided', 'less', 'greater').

    Returns:
        tuple: A tuple containing the Z-statistic and the p-value.
    """

    sample_mean = np.mean(sample_data)
    sample_size = len(sample_data)
    standard_error = population_std / np.sqrt(sample_size)
    z_statistic = (sample_mean - population_mean) / standard_error

    if alternative == 'two-sided':
        p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))
    elif alternative == 'less':
        p_value = stats.norm.cdf(z_statistic)
    elif alternative == 'greater':
        p_value = 1 - stats.norm.cdf(z_statistic)
    else:
        raise ValueError("Invalid alternative hypothesis. Choose from 'two-sided', 'less', or 'greater'.")

    print(f"Z-statistic: {z_statistic:.4f}")
    print(f"P-value: {p_value:.4f}")

    if p_value < alpha:
        print("Reject the null hypothesis.")
    else:
        print("Fail to reject the null hypothesis.")

    return z_statistic, p_value

# Example usage:
try:
    sample_data = [68, 72, 75, 63, 70, 78, 65, 71, 74, 69]
    population_mean = 70
    population_std = 5
    alpha = 0.05

    print("\nTwo-sided test:")
    z_test_hypothesis(sample_data, population_mean, population_std, alpha)

    print("\nLess than test:")
    z_test_hypothesis(sample_data, population_mean, population_std, alpha, alternative='less')

    print("\nGreater than test:")
    z_test_hypothesis(sample_data, population_mean, population_std, alpha, alternative='greater')

    z_test_hypothesis(sample_data, population_mean, population_std, alpha, alternative='bad_alternative') #example error.

except ValueError as e:
    print(f"Error: {e}")
```

```python
#20 Create a confidence interval for a dataset using Python and interpret the result.

import numpy as np
import scipy.stats as stats

def calculate_confidence_interval(data, confidence_level=0.95):
    """
    Calculates the confidence interval for a dataset.

    Args:
        data (list or numpy.ndarray): The dataset.
        confidence_level (float): The desired confidence level (default: 0.95).

    Returns:
        tuple: A tuple containing the lower and upper bounds of the confidence interval.
    """

    if not data:
        raise ValueError("Data cannot be empty.")

    data_array = np.array(data)
    sample_mean = np.mean(data_array)
    sample_std = np.std(data_array, ddof=1)  # Use ddof=1 for sample standard deviation
    sample_size = len(data_array)
    standard_error = sample_std / np.sqrt(sample_size)

    # Calculate the margin of error
    degrees_of_freedom = sample_size - 1
    margin_of_error = stats.t.ppf((1 + confidence_level) / 2, degrees_of_freedom) * standard_error

    # Calculate the confidence interval
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound

# Example usage:
try:
    data = [25, 30, 28, 35, 40, 32, 29, 33, 38, 31]
    confidence_level = 0.95

    lower, upper = calculate_confidence_interval(data, confidence_level)
    print(f"Confidence Interval ({confidence_level * 100}%): ({lower:.2f}, {upper:.2f})")

    #Interpretation
    print(f"We are {confidence_level * 100}% confident that the true population mean lies within the interval ({lower:.2f}, {upper:.2f}).")

    data2 = []
    lower, upper = calculate_confidence_interval(data2, confidence_level) #example error.

except ValueError as e:
    print(f"Error: {e}")
```

```python
#21 Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.

import numpy as np
import scipy.stats as stats

def generate_normal_and_confidence_interval(mu, sigma, sample_size, confidence_level=0.95):
    """
    Generates data from a normal distribution, calculates the confidence interval for the mean, and interprets the result.

    Args:
        mu (float): The population mean.
        sigma (float): The population standard deviation.
        sample_size (int): The size of the sample to generate.
        confidence_level (float): The desired confidence level (default: 0.95).
    """
    try:
        if sigma <= 0:
            raise ValueError("Standard deviation (sigma) must be positive.")
        if sample_size <= 0:
            raise ValueError("Sample size must be positive.")

        # Generate data
        data = np.random.normal(mu, sigma, sample_size)

        # Calculate confidence interval
        sample_mean = np.mean(data)
        sample_std = np.std(data, ddof=1)  # Sample standard deviation
        standard_error = sample_std / np.sqrt(sample_size)

        degrees_of_freedom = sample_size - 1
        margin_of_error = stats.t.ppf((1 + confidence_level) / 2, degrees_of_freedom) * standard_error

        lower_bound = sample_mean - margin_of_error
        upper_bound = sample_mean + margin_of_error

        print(f"Generated Data Mean: {sample_mean:.4f}")
        print(f"Confidence Interval ({confidence_level * 100}%): ({lower_bound:.4f}, {upper_bound:.4f})")

        # Interpretation
        print(f"Interpretation: We are {confidence_level * 100}% confident that the true population mean lies within the interval ({lower_bound:.4f}, {upper_bound:.4f}).")

        return (lower_bound, upper_bound)

    except ValueError as e:
        print(f"Error: {e}")
        return None

# Example usage:
try:
    mu = 50
    sigma = 10
    sample_size = 30
    confidence_level = 0.95

    generate_normal_and_confidence_interval(mu, sigma, sample_size, confidence_level)

    generate_normal_and_confidence_interval(mu, sigma, -5, confidence_level) # example error.
    generate_normal_and_confidence_interval(mu, -10, sample_size, confidence_level) # example error.
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
#22 Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution.

import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

def visualize_normal_pdf(mu, sigma, x_range=(-4, 4), num_points=1000):
    """
    Calculates and visualizes the probability density function (PDF) of a normal distribution.

    Args:
        mu (float): The mean of the normal distribution.
        sigma (float): The standard deviation of the normal distribution.
        x_range (tuple): The range of x-values to plot (in standard deviations).
        num_points (int): The number of points to generate for the plot.
    """
    try:
        if sigma <= 0:
            raise ValueError("Standard deviation (sigma) must be positive.")

        x = np.linspace(mu + x_range[0] * sigma, mu + x_range[1] * sigma, num_points)
        pdf = stats.norm.pdf(x, mu, sigma)

        plt.figure(figsize=(8, 6))
        plt.plot(x, pdf, 'b-', linewidth=2)
        plt.title(f"Normal Distribution PDF (μ={mu}, σ={sigma})")
        plt.xlabel("Values")
        plt.ylabel("Probability Density")
        plt.grid(True, linestyle='--')
        plt.show()

    except ValueError as e:
        print(f"Error: {e}")

# Example usage:
try:
    visualize_normal_pdf(0, 1)  # Standard normal distribution
    visualize_normal_pdf(5, 2)  # Normal distribution with mean 5, standard deviation 2
    visualize_normal_pdf(-3, 1.5) # normal distribution with negative mean.
    visualize_normal_pdf(0, -1) # example error.
except ValueError as e:
    print(f"Error: {e}")
```

```python
#23 Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def poisson_cdf_interpretation(lambda_val, x_values):
    """
    Calculates and interprets the cumulative distribution function (CDF) of a Poisson distribution.

    Args:
        lambda_val (float or int): The lambda (average rate) parameter of the Poisson distribution.
        x_values (list or numpy.ndarray): The values for which to calculate the CDF.
    """
    try:
        if lambda_val <= 0:
            raise ValueError("Lambda must be a positive number.")

        x_values = np.array(x_values)
        cdf_values = stats.poisson.cdf(x_values, lambda_val)

        print(f"Poisson CDF (λ={lambda_val}):")
        for x, cdf in zip(x_values, cdf_values):
            print(f"P(X <= {x}) = {cdf:.4f}")

        # Visualization
        plt.step(x_values, cdf_values, where='post')
        plt.xlabel("Number of Events (x)")
        plt.ylabel("Cumulative Probability (P(X <= x))")
        plt.title(f"Poisson CDF (λ={lambda_val})")
        plt.grid(True, linestyle='--')
        plt.show()

        # Interpretation example (for a specific x value)
        if len(x_values) > 0:
            example_x = x_values[0] #Take the first x_value from the list.
            example_cdf = cdf_values[0]
            print(f"\nInterpretation: There is a {example_cdf * 100:.2f}% probability that the number of events is less than or equal to {example_x}.")

        return cdf_values

    except ValueError as e:
        print(f"Error: {e}")
        return None

# Example usage:
try:
    lambda_val = 3.5
    x_values = [0, 1, 2, 3, 4, 5, 6]
    poisson_cdf_interpretation(lambda_val, x_values)

    lambda_val = -1 #example error.
    x_values = [0, 1, 2, 3, 4, 5, 6]
    poisson_cdf_interpretation(lambda_val, x_values)

except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
#24 Simulate a random variable using a continuous uniform distribution and calculate its expected value.

import numpy as np

def simulate_uniform_and_expected_value(a, b, num_samples=10000):
    """
    Simulates a random variable from a continuous uniform distribution and calculates its expected value.

    Args:
        a (float): The lower bound of the distribution.
        b (float): The upper bound of the distribution.
        num_samples (int): The number of samples to generate.

    Returns:
        tuple: A tuple containing the simulated samples and the calculated expected value.
    """
    try:
        if a >= b:
            raise ValueError("a must be less than b.")

        samples = np.random.uniform(a, b, num_samples)

        # Calculate the expected value (mean)
        expected_value = (a + b) / 2.0

        print(f"Simulated {num_samples} samples from Uniform({a}, {b})")
        print(f"Calculated Expected Value: {expected_value:.4f}")
        print(f"Sample Mean: {np.mean(samples):.4f}") #compare to sample mean.

        return samples, expected_value

    except ValueError as e:
        print(f"Error: {e}")
        return None, None

# Example usage:
try:
    a = 2.5
    b = 7.5
    samples, expected = simulate_uniform_and_expected_value(a, b)

    a = 10
    b = 5 #example error.
    samples, expected = simulate_uniform_and_expected_value(a, b)

except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

```python
#25 Write a Python program to compare the standard deviations of two datasets and visualize the difference.

import matplotlib.pyplot as plt
import numpy as np

def compare_standard_deviations(data1, data2, data1_label="Dataset 1", data2_label="Dataset 2"):
    """
    Compares the standard deviations of two datasets and visualizes the difference.

    Args:
        data1 (list or numpy.ndarray): The first dataset.
        data2 (list or numpy.ndarray): The second dataset.
        data1_label (str): Label for the first dataset (default: "Dataset 1").
        data2_label (str): Label for the second dataset (default: "Dataset 2").
    """
    try:
        if not data1 or not data2:
            raise ValueError("Datasets cannot be empty.")

        std_dev1 = np.std(data1)
        std_dev2 = np.std(data2)

        print(f"Standard Deviation of {data1_label}: {std_dev1:.4f}")
        print(f"Standard Deviation of {data2_label}: {std_dev2:.4f}")

        # Visualization
        plt.figure(figsize=(8, 6))
        plt.bar([data1_label, data2_label], [std_dev1, std_dev2], color=['skyblue', 'lightcoral'])
        plt.ylabel("Standard Deviation")
        plt.title("Comparison of Standard Deviations")
        plt.grid(True, axis='y', linestyle='--')
        plt.show()

        return std_dev1, std_dev2

    except ValueError as e:
        print(f"Error: {e}")
        return None, None

# Example usage:
try:
    data1 = [25, 30, 28, 35, 40, 32, 29, 33, 38, 31]
    data2 = [15, 20, 18, 25, 30, 22, 19, 23, 28, 21]

    compare_standard_deviations(data1, data2, "Group A", "Group B")

    data3 = []
    compare_standard_deviations(data1, data3, "Group A", "Group C") #example error.
except ValueError as e:
    print(f"Error: {e}")
```

```python
#26 Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution.

import numpy as np
import scipy.stats as stats

def calculate_range_and_iqr_normal(mu, sigma, sample_size=1000):
    """
    Generates data from a normal distribution and calculates its range and interquartile range (IQR).

    Args:
        mu (float): The mean of the normal distribution.
        sigma (float): The standard deviation of the normal distribution.
        sample_size (int): The number of samples to generate.

    Returns:
        tuple: A tuple containing the range and IQR.
    """
    try:
        if sigma <= 0:
            raise ValueError("Standard deviation (sigma) must be positive.")

        data = np.random.normal(mu, sigma, sample_size)

        # Calculate range
        data_range = np.max(data) - np.min(data)

        # Calculate IQR
        q75, q25 = np.percentile(data, [75, 25])
        iqr = q75 - q25

        print(f"Generated Normal Distribution (μ={mu}, σ={sigma})")
        print(f"Range: {data_range:.4f}")
        print(f"IQR: {iqr:.4f}")

        return data_range, iqr

    except ValueError as e:
        print(f"Error: {e}")
        return None, None

# Example usage:
try:
    mu = 50
    sigma = 10
    data_range, iqr = calculate_range_and_iqr_normal(mu, sigma)

    mu = 50
    sigma = -10 #example error.
    data_range, iqr = calculate_range_and_iqr_normal(mu, sigma)

except ValueError as e:
    print(f"Error: {e}")
```

```python
#27 Implement Z-score normalization on a dataset and visualize its transformation.

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

def z_score_normalize_and_visualize(data, data_label="Original Data"):
    """
    Performs Z-score normalization on a dataset and visualizes the transformation.

    Args:
        data (list or numpy.ndarray): The dataset.
        data_label (str): Label for the original dataset (default: "Original Data").
    """
    try:
        if not data:
            raise ValueError("Data cannot be empty.")

        data_array = np.array(data)
        mean = np.mean(data_array)
        std_dev = np.std(data_array)

        if std_dev == 0:
            raise ValueError("Standard deviation is zero. Z-scores cannot be calculated.")

        z_scores = (data_array - mean) / std_dev

        # Visualization
        plt.figure(figsize=(12, 6))

        # Original data plot
        plt.subplot(1, 2, 1)
        sns.histplot(data_array, kde=True)
        plt.title(f"{data_label} Distribution")
        plt.xlabel("Values")
        plt.ylabel("Frequency")
        plt.grid(True, linestyle='--')

        # Z-score normalized data plot
        plt.subplot(1, 2, 2)
        sns.histplot(z_scores, kde=True, color='orange')
        plt.title("Z-score Normalized Distribution")
        plt.xlabel("Z-scores")
        plt.ylabel("Frequency")
        plt.grid(True, linestyle='--')

        plt.tight_layout()
        plt.show()

        return z_scores

    except ValueError as e:
        print(f"Error: {e}")
        return None

# Example usage:
try:
    data = [10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60]
    z_scores = z_score_normalize_and_visualize(data, "Sample Data")

    data2 = [5,5,5,5,5]
    z_scores2 = z_score_normalize_and_visualize(data2, "Constant Data") # example error.
    data3 = []
    z_scores3 = z_score_normalize_and_visualize(data3, "Empty Data") # example error.


except ValueError as e:
    print(f"Error: {e}")
```

```python
#28 Q28. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
import scipy.stats as stats

def calculate_skewness_kurtosis_normal(mu, sigma, sample_size=1000):
    """
    Generates data from a normal distribution and calculates its skewness and kurtosis.

    Args:
        mu (float): The mean of the normal distribution.
        sigma (float): The standard deviation of the normal distribution.
        sample_size (int): The number of samples to generate.

    Returns:
        tuple: A tuple containing the skewness and kurtosis.
    """
    try:
        if sigma <= 0:
            raise ValueError("Standard deviation (sigma) must be positive.")

        data = np.random.normal(mu, sigma, sample_size)

        # Calculate skewness and kurtosis
        skewness = stats.skew(data)
        kurtosis = stats.kurtosis(data)

        print(f"Generated Normal Distribution (μ={mu}, σ={sigma})")
        print(f"Skewness: {skewness:.4f}")
        print(f"Kurtosis: {kurtosis:.4f}")

        return skewness, kurtosis

    except ValueError as e:
        print(f"Error: {e}")
        return None, None

# Example usage:
try:
    mu = 50
    sigma = 10
    skew, kurt = calculate_skewness_kurtosis_normal(mu, sigma)

    mu = 50
    sigma = -10 #example error.
    skew, kurt = calculate_skewness_kurtosis_normal(mu, sigma)

except ValueError as e:
    print(f"Error: {e}")
```
